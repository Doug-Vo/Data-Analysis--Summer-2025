```python
df = pd.read_csv('/kaggle/input/ai-vs-human-content-detection-1000-record-in-2025/ai_human_content_detection_dataset.csv')



df.info()
df.isnull().sum()
```

```python
processed_df = df.drop(columns = ['text_content'], axis = 1)

processed_df = processed_df.dropna()
processed_df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler


le = LabelEncoder()

processed_df['content_type'] = le.fit_transform(processed_df['content_type'])

X = processed_df.drop(columns =['label'], axis = 1)
y = processed_df['label']

sns.countplot(x = y, data = processed_df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)

log_model = LogisticRegression(random_state = 42, max_iter = 1000)

log_model.fit(X_train, y_train)

y_pred = log_model.predict(X_test)

print(classification_report(y_pred, y_test))

cm = confusion_matrix(y_pred, y_test)

sns.heatmap(cm, fmt = 'g', annot = True, cmap = 'Blues')
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier

mlp_model = MLPClassifier(
    random_state = 42, 
    hidden_layer_sizes = (100, 200),
    early_stopping = True,
)


mlp_model.fit(X_train, y_train)

mlp_y_pred = mlp_model.predict(X_test)


```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

X_text = df['text_content']
y_text = df['label']


X_train, X_test, y_train, y_test = train_test_split(X_text, y_text, test_size=0.2, random_state=42, stratify=y_text)
```

```python
vectorizer = TfidfVectorizer(max_features = 5000, stop_words = 'english')

X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier



model = RandomForestClassifier(random_state = 42, n_estimators = 100, n_jobs = -1)

model.fit(X_train_vectorized, y_train)

y_pred = model.predict(X_test_vectorized)

print(classification_report(y_pred, y_test))

cm = confusion_matrix(y_pred, y_test)

sns.heatmap(cm, fmt = 'g', annot = True, cmap = 'Blues')
plt.show()
```

```python
from sklearn.svm import LinearSVC



model = LinearSVC(random_state = 42)

model.fit(X_train_vectorized, y_train)

y_pred = model.predict(X_test_vectorized)

print(classification_report(y_pred, y_test))

cm = confusion_matrix(y_pred, y_test)

sns.heatmap(cm, fmt = 'g', annot = True, cmap = 'Blues')
plt.show()
```

### Conclusion: The Challenge of Differentiating AI and Human Text

After a thorough analysis using multiple machine learning models, we have arrived at a significant conclusion regarding the difficulty of this classification task.

Several models were trained and evaluated, including:
* `LogisticRegression`
* `MLPClassifier`
* `RandomForestClassifier`
* `LinearSVC`

Furthermore, we enhanced our text preprocessing by using **TF-IDF Vectorization**, including both single words (unigrams) and pairs of words (bigrams), to help the models capture more context.

Despite these efforts, all of the models consistently produced an **accuracy of approximately 50%**. In a binary classification problem like this one (AI vs. Human), a 50% accuracy is equivalent to random guessing. This indicates that the models were unable to find any reliable patterns in the word or phrase frequencies to distinguish between the two classes.

This outcome suggests that modern AI-generated text has become so sophisticated that it is extremely difficult, and perhaps impossible, for these traditional machine learning models to differentiate it from human-written content based on text features alone.
