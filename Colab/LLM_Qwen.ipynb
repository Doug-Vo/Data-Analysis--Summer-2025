```python
from transformers import AutoTokenizer
from transformers import AutoModelForCausalLM
import torch
from transformers import set_seed
import warnings
warnings.filterwarnings("ignore")
```

```python
model_name = "Qwen/Qwen2-0.5B"

# load the tokenizer and the model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)



prompt = "I am very tired"
```

```python
inputs_id = tokenizer(prompt, return_tensors = 'pt').input_ids.to("cuda")

outputs = model(inputs_id)

```

```python
set_seed(42)

outputs_id = model.generate(inputs_id, max_new_tokens = 20)

decoded_text = tokenizer.decode(outputs_id[0])

print(decoded_text)


```

```python
beam_output = model.generate(
    inputs_id,
    num_beams = 5,
    max_new_tokens = 20
)


print(tokenizer.decode(beam_output[0]))
```

```python
output = model.generate(
    inputs_id,
    num_beams = 8,
    max_new_tokens = 30,
    temperature = 1.2,
    repetition_penalty = 1.2
    )

print(tokenizer.decode(output[0]))
```
